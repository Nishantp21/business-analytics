```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as st
```

```python
df=pd.read_excel("/Regression Data.xlsx")
```

```python
# df
```

```python
volume = df.Volume
# volume
```

```python
weight = df.Weight
# weight
```

```python
co2 = df.CO2
# co2
```

```python
slope, intercept, r, p, std_err = st.linregress(volume, co2)

def y(x):
  return slope * x + intercept

mymodel = list(map(y, volume))

print("r value for the this model is ",r)

plt.scatter(volume, co2)
plt.plot(volume, mymodel)
plt.show()
```

```python
slope, intercept, r, p, std_err = st.linregress(weight, co2)

def y(x):
  return slope * x + intercept

mymodel = list(map(y, weight))

print("r value for the this model is ",r)

plt.scatter(weight, co2)
plt.plot(weight, mymodel)
plt.show()
```

```python
from sklearn import linear_model as lm
```

```python
X = df[['Weight' , 'Volume']]
# X
```

```python
y = df['CO2']
# y
```

```python
regr = lm.LinearRegression()
regr.fit(X, y)
```

```python
(weight_coef, volume_coef) = regr.coef_
print("Weight Coefficient is : ", weight_coef)
print("Volume Coefficient is : ", volume_coef)
```

```python
input_weight = input("Enter the Weight : ")
input_volume = input("Enter the Volume : ")
```

```python
co2predict = regr.predict([[input_weight, input_volume]])
print("Predicted CO2 from this Model is : ",co2predict)
```
