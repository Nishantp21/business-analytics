```python
X = df.label
X
```

```python
y = np.array(df.tempo).reshape(-1,1)
y
```

```python
music = lm.LogisticRegression()
music.fit(y,X)
```

```python
to_predict = input("Enter the data : ")
predicted = music.predict(np.array([to_predict]).reshape(-1,1))
print(predicted)
```

```python
def music2prob(logr, X):
  log_odds = logr.coef_ * X + logr.intercept_
  odds = np.exp(log_odds)
  probability = odds / (1 + odds)
  return(probability)
```

```python
print(music2prob(music, y))
```
